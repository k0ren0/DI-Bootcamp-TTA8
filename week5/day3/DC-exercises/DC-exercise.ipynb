## Task 1:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
```

```python
y = x**2
```

```python
plt.plot(x, y, label='y = x^2')
plt.show()
```

```python
plt.plot(x, y)

plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(0, 20)
plt.ylim(0, 400)

plt.savefig('plot_task1.png')
plt.show()
```

## Task 2:

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
plt.scatter(chipo['item_price'], chipo['quantity'])
plt.title('Relationship between Items the Price and Quantity Ordered')
plt.xlabel('Item_price')
plt.ylabel('Quantity ordered')

plt.savefig('plot_task2.png')
plt.show()
```

## Task 3:

```python
chipo_grouped = chipo.groupby('item_name')['quantity'].sum().reset_index()
```

```python
top5_items = chipo_grouped.sort_values(by='quantity', ascending=False).head(5)
```

```python
plt.bar(top5_items['item_name'], top5_items['quantity'])
plt.title('Top 5 Items by Quantity Ordered')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')

plt.savefig('plot_task3.png')
plt.show()
```
