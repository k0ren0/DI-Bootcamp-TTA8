## ðŸŒŸ Exercise 7 :

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, y1, color='tab:blue', label='y = sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = sin(x)')
axes[0].legend()

axes[1].scatter(x, y2, marker='o', color='tab:blue', label='y = cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('y = cos(x)')
axes[1].legend()

plt.tight_layout()
plt.show()
```
