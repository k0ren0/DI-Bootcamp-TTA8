```python
oil = pd.read_csv("./oil.csv")
oil
```

```python
oil.plot(x="date")
```

```python
(oil
.astype({"date":"datetime64[ns]"})
.set_index("date")
.loc["2013":"2014"]
.plot())
```

```python
oil.plot(
    x="date",
    title="oil transactions")
```

```python
oil.plot(
    x="date",
    title="Transactions",
    ylabel="date",
)
```

```python
transactions.plot(
    x="date",
    title="Transactions",
    ylabel="date",
    color=["pink", "blue"]
)
```

```python
transactions.plot(
    x="date",
    title="Transactions",
    ylabel="date",
    cmap='Pastel1'
)
```

```python
transactions.plot.pie(
    y="transactions",
    x="date",
    title="Transactions",
    ylabel="date",
    cmap='Pastel1')
```

```python
transactions.plot.scatter(
    y="transactions",
    x="date"
)
```
